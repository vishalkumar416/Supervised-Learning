```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
```

```python
# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([10000, 20000, 30000, 40000, 50000])  # Salary
```

```python
# Create model
model = DecisionTreeRegressor()
```

```python
# Train model
model.fit(X, y)
# Predict
prediction = model.predict([[3]])
print("Predicted Salary:", prediction)
```
