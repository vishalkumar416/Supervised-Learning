```python
from sklearn.svm import SVR
import numpy as np

# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([10000, 20000, 30000, 40000, 50000])

# Create model
model = SVR(kernel='linear')

# Train model
model.fit(X, y)

# Predict
prediction = model.predict([[3.5]])
print("Predicted Value:", prediction)
```
